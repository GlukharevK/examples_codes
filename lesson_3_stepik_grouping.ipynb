```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_performance.head()
```

```python
students_performance.columns = students_performance.columns.str.replace(" ", "_")
```

## Группировка + aggregate ~ признак по которому группирует

```python
students_performance.groupby('gender').mean()
```

```python
students_performance.groupby('gender', as_index=False) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

```python
students_performance.groupby('gender', as_index=False) \
    .aggregate({'math_score' : ['mean'], 'reading_score' : ['mean']})
```

```python
students_performance.groupby(['gender', 'race/ethnicity'], as_index=False) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

### без индексов:

```python
students_performance.groupby(['gender', 'race/ethnicity']) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

## мультииндексы:

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']) \
    .aggregate({'math_score' : 'mean', 'reading_score' : 'mean'}) \
    .rename(columns={'math_score' : 'mean_math_score', 'reading_score' : 'mean_reading_score'})
```

## Создание новых колонок:

```python
students_performance['total_score'] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

### или:

```python
students_performance.filter(like='score').sum(axis=1)
```

```python
students_performance.head()
```

### добавление с помощью метода assign:

```python
students_performance = students_performance.assign(total_score_log = np.log(students_performance.total_score))
```

```python
students_performance.head()
```

### Удаление колонок

```python
students_performance = students_performance.drop(['total_score'], axis=1)
```

### Пользуясь предыдущими данными, укажите через пробел (без запятых) чему равны минимальная, средняя и максимальная концентрации аланина (alanin) среди видов рода Fucus. Округлите до 2-ого знака, десятичным разделителем является точка.

```python
seaweed_df.query('genus == "Fucus"').agg(['min', 'mean', 'max']).filter(['alanin']).round(2)
```

```python
seaweed_df.groupby('genus').agg(['min', 'mean', 'max']).loc[['Fucus'], ['alanin']].round(2)
```

```python
seaweed_df.groupby('genus').agg({'alanin' : ['min', 'mean', 'max']}).loc['Fucus'].unstack().round(2)
```

### Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях

```python
seaweed_df.groupby('group').agg({'citrate' : 'var'}).round(2)
```

### Число видов бурых, зелёных и красных водорослей

```python
seaweed_df.value_counts('group')
```

### Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях

```python
seaweed_df.groupby('group').agg({'sucrose' : lambda x : max(x) - min(x)}).round(2)
```

### Все вместе:

```python
seaweed_df.groupby('group').aggregate({'sucrose' : ['count', 'var', np.ptp]}).round(2)
```
