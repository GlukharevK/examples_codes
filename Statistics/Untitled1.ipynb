```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel('orders.xls')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4'], axis=1)
```

```python
df = df[:-1]
```

```python
df['Заказ клиента, Клиент, Пометка удаления'].values
```

```python
df['Заказ клиента, Клиент, Пометка удаления'] = df['Заказ клиента, Клиент, Пометка удаления'].astype(str)
```

```python
row1 = df['Заказ клиента, Клиент, Пометка удаления']
```

```python
row1 = list(row1.values)
```

```python
row1[0]
```

```python
pattern = r"(\d{6}).*?(\d{2}.\d{2}.\d{4}) \d*:\d*:\d*, (.*), (.*)"
```

```python
res = []
```

```python
for i in range(len(row1)):
    print(re.findall(pattern, row1[i])[0])
    res.append(re.findall(pattern, row1[i])[0])
```

```python
order_client = []
date = []
dealer = []
delete_note = []
```

```python
for j in res:
    order_client.append(j[0])
    date.append(j[1])
    dealer.append(j[2])
    delete_note.append(j[3])
```

```python
pd.Series(order_client)
```

```python
new_df = pd.DataFrame()
```

```python
new_df['Заказ клиента'] = pd.Series(order_client)
new_df['Дата заказа'] = pd.Series(date)
new_df['Дилер'] = pd.Series(dealer)
new_df['Пометка удаления'] = pd.Series(delete_note)
```

```python
new_df
```

```python
new_df.to_excel('new_data.xlsx')
```

```python
new_df.to_csv('new_data.csv')
```
