```python
from sklearn import tree
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
cat_dog_df = pd.read_csv('https://stepik.org/media/attachments/course/4852/dogs_n_cats.csv')
```

```python
cat_dog_df
```

```python
cat_dog_df.isna().sum()
```

```python
cat_dog_df = pd.get_dummies(cat_dog_df)
```

```python
cat_dog_df
```

```python
cat_dog_df = cat_dog_df.drop('Вид_котик', axis=1).rename(columns={'Вид_собачка' : 'Вид'})
```

```python
cat_dog_df
```

```python
X_train = cat_dog_df.drop('Вид', axis=1)
```

```python
y_train = cat_dog_df['Вид']
```

```python
max_depth_tree = range(1,100)
rs = np.random.seed(0)
scores_data = pd.DataFrame()
```

```python
for max_depth in max_depth_tree:
    clf = tree.DecisionTreeClassifier(criterion='entropy', random_state=rs)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    
    temp_score_data = pd.DataFrame({'max_depth' : [max_depth],
                                    'train_score' : [train_score],
                                    'mean_cross_val_score' : mean_cross_val_score})
    scores_data = scores_data.append(temp_score_data)
```
