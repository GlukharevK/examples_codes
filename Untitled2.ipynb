```python
from collections import defaultdict
```

```python
with open('C:/Users/glukh/JupyterFiles/numbers.txt') as file:
    k_lst = [int(x) for x in file.readline().split(',')]
```

```python
k_lst
```

```python
cache = defaultdict(int)

def fibonacci(num):
    if num in [0, 1, 2]:
        cache[num] = num
        return num
    if cache[num]:
        return cache[num]
    else:
        res = fibonacci(num - 1) + fibonacci(num - 3)
        cache[num] = res
        return res

fibonacci(max(k_lst))
print(' '.join(str(cache[i]) for i in k_lst))
```
