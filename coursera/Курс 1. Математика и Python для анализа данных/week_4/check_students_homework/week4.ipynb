```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

 Берем вот это распределение: https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.semicircular.html 

```python
rv = sts.semicircular()
```

```python
sc = sts.semicircular()
```

```python
r = sc.rvs(1000)
```

```python
x = np.linspace(-1,1,1000)
```

```python
plt.hist(r,normed=True, histtype='stepfilled', alpha=0.2)
plt.plot(x,sc.pdf(x))
```

Теоретические значения мат ожидания и среднеквадратического отклонения берем из вики:
    loc=0, scale =1/2

```python
norm__ = sts.norm(loc =0,scale=1/2)
```

```python
plt.hist(r, normed=True, histtype='stepfilled',bins=30,alpha=0.5)
plt.plot(x,norm__.pdf(x))
```

Генерируем 1000 выборок объема 5,20,100 и строим массивы выборочных средних

```python
arr5=[]
arr20=[]
arr100=[]
```

```python
for i in range(1000):
    rv5 = sc.rvs(5)
    arr5.append(rv5.mean())
for i in range(1000):
    rv20 = sc.rvs(10)
    arr20.append(rv20.mean())
for i in range(1000):
    rv100 = sc.rvs(100)
    arr100.append(rv100.mean())
```

$$Параметры__нормального распределения по средневыборочному$$
$$ loc= mu, $$ $$ scale = \sqrt{Dx/n}$$
$$ mu = 0 , Dx = 1/4 $$ 
$$n = 5: loc = 0,  scale = \sqrt{1/20} $$
$$n = 20: loc = 0,  scale = \sqrt{1/80}$$ 
$$n = 100: loc = 0,  scale = \sqrt{1/400}$$ 

```python
norm5 = sts.norm(loc =0,scale=1/4.47)
```

```python
norm20 = sts.norm(loc =0,scale=1/8.94)
```

```python
norm100 = sts.norm(loc =0,scale=1/20)
```

```python
plt.hist(arr5,normed=True, histtype='stepfilled',label = 'n = 5',bins=30,alpha=0.5)
plt.plot(x,norm5.pdf(x))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
```

```python
plt.hist(arr20,normed=True, histtype='stepfilled',label = 'n = 20',bins=40,alpha=0.5)
plt.plot(x,norm20.pdf(x))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
```

```python
plt.hist(arr100,normed=True, histtype='stepfilled',label = 'n = 100',alpha=0.5)
plt.plot(x,norm100.pdf(x))
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
```

$$__Вывод: распределение хорошо апроксимируется нормальным$$
