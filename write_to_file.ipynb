```python
with open('numbers.txt', 'w') as file:
    file.writelines(str(x) + '\n' for x in lst)
```

```python
with open('numbers.txt', 'wb') as file:
    pickle.dump(lst, file)
```

```python
with open('numbers.txt', 'w') as file:
    file.write(','.join(str(x) for x in lst))
```

```python
with open('numbers_sessions.txt', 'w') as file:
    file.write(','.join(str(x) for x in lst))
    for i in range(1000):
        np.random.randint(0, 1000000000, 1000)
```
